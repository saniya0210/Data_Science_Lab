This program will identify whether a given integer, entered by the user, is an odd or an even number.

```python
def check_odd_even(number):
    """
    Checks if a given number is odd or even.

    Args:
        number (int): The integer to check.

    Returns:
        str: 'Even' if the number is even, 'Odd' if the number is odd.
    """
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Take input from the user
try:
    user_input = int(input("Enter an integer to check if it's odd or even: "))
    result = check_odd_even(user_input)
    print(f"The number {user_input} is {result}.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Take input from the user
num_str = input("Enter a non-negative integer to compute its factorial: ")

try:
    num = int(num_str)
    result = factorial(num)
    print(f"The factorial of {num} is {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
# Take input from the user
num_str = input("Enter an integer to display its multiplication table: ")

try:
    num = int(num_str)
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):  # Display table up to 10
        print(f"{num} x {i} = {num * i}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
# Take input from the user
num_str = input("Enter an integer to check if it's odd or even: ")

try:
    num = int(num_str)
    if num % 2 == 0:
        print(f"The number {num} is even.")
    else:
        print(f"The number {num} is odd.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
# Take input from the user
num_str = input("Enter an integer to check if it's a palindrome: ")

try:
    num = int(num_str)
    # Convert the number to a string to easily reverse it
    s = str(num)
    # Check if the string is equal to its reverse
    if s == s[::-1]:
        print(f"The number {num} is a palindrome.")
    else:
        print(f"The number {num} is not a palindrome.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```
